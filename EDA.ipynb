### Maps
Trong pandas, "map" đề cập đến một hàm thực hiện việc lấy một tập hợp giá trị và "ánh xạ" chúng thành một tập hợp giá trị khác. Trong khoa học dữ liệu, chúng ta thường cần tạo các biểu diễn mới từ dữ liệu hiện có hoặc chuyển đổi dữ liệu từ định dạng hiện tại sang định dạng mà chúng ta muốn sử dụng sau này.

Trong pandas, có hai phương thức ánh xạ (mapping) thường sử dụng:

Phương thức map(): Đây là phương thức đầu tiên và đơn giản.
Khi sử dụng phương thức map(), bạn có thể ánh xạ các giá trị trong một chuỗi hoặc cột sang một tập hợp giá trị khác dựa trên một ánh xạ được xác định trước. Ánh xạ này có thể là một hàm, một từ điển hoặc một Series. Điều này cho phép bạn thực hiện các biến đổi dữ liệu đơn giản mà không cần thay đổi cấu trúc của DataFrame ban đầu.

Ví dụ, nếu bạn có một cột chứa các giá trị "nam" và "nữ" và muốn chuyển chúng thành 0 và 1 tương ứng, bạn có thể sử dụng phương thức map() để thực hiện việc này.

````python
SepalLengthMean = df_iris.SepalLengthCm.mean()
df_iris.SepalLengthCm.map(lambda x: x - SepalLengthMean)
````

Hàm **map()** sẽ không làm thay đổi giá trị của DataFrame mà chỉ trả về giá trị Series đã được ánh xạ. Muốn thay đổi cả DataFrame ta cần dùng hàm **apply()**. Ví dụ:

````python
def remean_sepal(row):
    row.SepalLengthCm = row.SepalLengthCm - SepalLengthMean
    return row

df_iris.apply(remean_sepal, axis='columns')
````

## Grouping, Sorting

### Groupwise analysis

````python
download_url = (
    "https://raw.githubusercontent.com/fivethirtyeight/data/master/college-majors/recent-grads.csv"
)
df_major = pd.read_csv(download_url)
df_major
````

````python
df_major.groupby('Major_category')
````

**groupby()** có thể sử dụng để xem thông tin của 1 hay nhiều cột và có thể sử dụng cùng các hàm khác như **count(), min(), max(),...** Ví dụ như dưới đây:

````python
df_major.groupby('Major_category').Major_category.count() # ~ df_major.Major_category.value_counts()
````

Mỗi group chúng ta tạo ra có thể hiểu là một phần được cắt từ DataFrame mà chứa dữ liệu về giá trị chúng ta muốn. Có thể sử dụng **apply()** để thao tác theo cách mình muốn. Ví dụ như dưới đây:

````python
df_major.groupby('Major').apply(lambda df: df.Total.iloc[0])
````

Hàm **agg()** được sử dụng để thực hiện các thao tác tổng hợp hoặc tính toán trên dữ liệu trong một nhóm hoặc trên cả DataFrame.

Cú pháp của hàm agg() như sau:
```
DataFrame.agg(func, axis=0, *args, **kwargs)
```
Trong đó:

- **DataFrame** hoặc **Series** là các cấu trúc dữ liệu trong Pandas.
- **func** là một hàm hoặc danh sách các hàm được áp dụng để tính toán tổng hợp. Hàm có thể được xây dựng sẵn trong Python hoặc tự định nghĩa. Ví dụ, nếu bạn muốn tính trung bình của một nhóm dữ liệu, bạn có thể sử dụng hàm **mean(), min(), max(),...**.
- **axis** là đối số tùy chọn, có giá trị mặc định là 0 (áp dụng theo các cột). Nếu bạn muốn áp dụng **agg()** theo các hàng, có thể đặt **axis=1**.

Ví dụ:

````python
df_major.groupby(['Major_category']).Total.agg([len, min, max])
````

## Box plot

**Boxplot** là một dạng biểu đồ thể hiện phân phối dữ liệu của các thuộc tính số thông qua các “tứ phân vị”.Tứ phân vị là một khái niệm trong thống kê dùng để mô tả sự phân bố và sự phân tán của tập dữ liệu, gồm 3 giá trị: Q1, Q2 và Q3 chia tập dữ liệu thành 4 phần bằng nhau.

**Boxplot** thể hiện các phân phối dữ liệu, nghĩa là giúp chúng ta biết được độ dàn trải của các điểm dữ liệu như thế nào, dữ liệu có đối xứng không, phân bố rộng hay hẹp, giá trị nhỏ nhất, lớn nhất và các điểm ngoại lệ.
Biểu đồ Boxplot thể hiện 5 thông số:

- **Median**: Trung vị của tập dữ liệu, tức là giá trị ở phần tử giữa.
- **First quartile (Q1)**: Trung vị giữa Median và phần tử nhỏ nhất trong tập dữ liệu. Còn gọi là 25th Percentile.
- **Third quartile (Q3)**: Trung vị giữa Median và phần tử lớn nhất trong tập dữ liệu. Còn gọi là 75th Percentile.
- **Minimum**: Phần tử nhỏ nhất không phải ngoại lệ.
- **Maximum**: Phần tử lớn nhất không phải là ngoại lệ.

**Box plot** thường được sử dụng để kiểm chứng xem dữ liệu có lệch không và hiển thị các điểm dữ liệu ngoại lệ

# PCA

````python
from sklearn.preprocessing import StandardScaler
features = df_iris.columns[1:5]
x = df_iris.loc[:, features].values
x = StandardScaler().fit_transform(x)

from sklearn.decomposition import PCA, TruncatedSVD
pca_iris = PCA(n_components=2)
principalComponents_iris = pca_iris.fit_transform(x)
principalComponents_iris
````

````python
svd_iris = TruncatedSVD(n_components=2)
princComp_iris =  svd_iris.fit_transform(x)
princComp_iris
````

````python
principal_iris_df = pd.DataFrame(data = principalComponents_iris
             , columns = ['principal component 1', 'principal component 2'])
principal_iris_df.describe()
````

````python
princSVD_iris_df = pd.DataFrame(data = princComp_iris
             , columns = ['principal component 1', 'principal component 2'])
princSVD_iris_df.describe()
````

````python
plt.figure()
plt.figure(figsize=(5,5))
plt.xticks(fontsize=6)
plt.yticks(fontsize=7)
plt.xlabel('Principal Component - 1',fontsize=10)
plt.ylabel('Principal Component - 2',fontsize=10)
plt.title("Principal Component Analysis of Iris Dataset",fontsize=10)
targets = df_iris.Species.unique().tolist()
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = df_iris['Species'] == target
    plt.scatter(principal_iris_df.loc[indicesToKeep, 'principal component 1'],
                principal_iris_df.loc[indicesToKeep, 'principal component 2'], c = color, s = 20)

plt.legend(targets,prop={'size': 7})
````

````python
plt.figure()
plt.figure(figsize=(5,5))
plt.xticks(fontsize=6)
plt.yticks(fontsize=7)
plt.xlabel('Principal Component - 1',fontsize=10)
plt.ylabel('Principal Component - 2',fontsize=10)
plt.title("Principal Component Analysis of Iris Dataset",fontsize=10)
targets = df_iris.Species.unique().tolist()
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = df_iris['Species'] == target
    plt.scatter(princSVD_iris_df.loc[indicesToKeep, 'principal component 1'],
                princSVD_iris_df.loc[indicesToKeep, 'principal component 2'], c = color, s = 20)

plt.legend(targets,prop={'size': 7})
````

# Correlation Matrix

Cũng giống với ma trận hiệp phương sai, **Correlation Matrix (ma trận tương quan)** là một bảng thể hiện hệ số tương quan giữa các biến khi ta có nhiều hơn 2 biến trong bộ dữ liệu. Mỗi ô trong bảng hiển thị mối tương quan giữa hai biến. 

````python
sepal_petal_corr = df_iris["SepalLengthCm"].corr(df_iris["PetalLengthCm"]) 
print("The correlation number between Sepal length and Petal width: ", sepal_petal_corr)
````
